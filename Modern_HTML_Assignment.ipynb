1. List out the features of HTML5?
- i. Semantic Elements – New tags like ```<header>```, ```<footer>```, ```<article>```, and ```<section>``` make the structure more meaningful.

  ii. Multimedia Support – Supports ```<audio>``` and ```<video>``` tags to embed media without plugins like Flash.

  iii. Canvas and SVG – Allows drawing graphics and animations using ```<canvas>``` and ```<svg>```.

  iv. Responsive Design – Works better on different screen sizes with ```<meta viewport>``` and CSS media queries.

  v. Web Storage – Provides localStorage and sessionStorage to store data in the browser.

  vi. Better Performance – Faster loading and improved efficiency in running scripts.


2. What are HTML Entities? List out 5 commonly used HTML entities.
- HTML entities are special codes used to display reserved characters (like <, >, &, ", etc.) or symbols that cannot be typed directly in HTML. They start with "&" and end with ";".

  5 Commonly Used HTML Entities:
  i. ```&lt;``` → < (Less than)
  ii. ```&gt;``` → > (Greater than)
  iii. ```&amp;``` → & (Ampersand)
  iv. ```&quot;``` → " (Double Quote)
  v. ```&apos;``` → ' (Single Quote)


3. Define accessibility in the context of web development. Discuss why it's essential to create accessible websites and how it benefits different user group?
- Accessibility in web development means designing and developing websites in a way that everyone, including people with disabilities, can use them easily.

  Why is Accessibility Essential?
  - Inclusivity – Ensures that people with disabilities (visual, hearing, motor, or cognitive impairments) can access web content.
  - Better User Experience – Improves usability for all users, including elderly people and those with temporary disabilities.
  - SEO Benefits – Search engines prefer well-structured, accessible websites, improving rankings.

  How Accessibility Benefits Different User Groups?
  - Visually Impaired Users – Screen readers help them navigate websites if proper headings, alt texts, and ARIA labels are used.
  - Hearing-Impaired Users – Captions and transcripts for videos help them understand content.
  - Users with Motor Disabilities – Keyboard-friendly navigation and voice commands make browsing easier.


4. List any 3 ways which help us in improving the accessibility of HTML.
- Three Ways to Improve HTML Accessibility
  
  i. Use Semantic HTML:
    - Use proper tags like ```<header>```, ```<nav>```, ```<main>```, ```<section>```, ```<article>```, and ```<footer>``` to improve structure and meaning.
    - Helps screen readers and search engines understand the content better.
  
  ii. Provide Alternative Text for Images (alt Attribute) –
    -  Always add alt="description" for images so visually impaired users using screen readers can understand them.
    - Example: ```<img src="flower.jpg" alt="A red rose in a garden">```

  iii. Ensure Keyboard Navigation –
    - Make sure all elements (buttons, links, forms) can be accessed using the Tab key.
    - Use tabindex for better focus control.
    - Example: ```<button tabindex="0">Click Me</button>```
